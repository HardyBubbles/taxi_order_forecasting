Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.

Вам нужно:

1. Загрузить данные и выполнить их ресемплирование по одному часу.

## Обучение

### Подготовка данных

До разбиения датафрема на выборки нам нужно досоздавать признаки.  
Добавим календарные признаки, предыдущие (отстающие) и скользящее среднее.  
Размеры лага для отстающих признаков и размер окна для скользящего среднего мы определяем сами, от них будет зависеть качество данных и обучение моделей.

```python
def make_features(df_func, col_name, max_lag, rms):  # rms - rolling mean size
    #df_func['month'] = df_func.index.month
    #df_func['day'] = df_func.index.day
    df_func['dayofweek'] = df_func.index.dayofweek
    df_func['hour'] = df_func.index.hour
    
    for lag in range(1, max_lag+1):
        df_func['lag_{}'.format(lag)] = df_func[col_name].shift(lag)
    
    df_func['rolling_mean'] = df_func[col_name].shift().rolling(rms).mean()
    df_func.dropna(inplace=True)
    
    return df_func
```

```python
df_ml = df.copy()
df_ml = make_features(df_ml, 'num_orders', 24, 24)
# таким образои максимальный размер лага равен суткам и размер окна для скользящего среднего тоже
```

```python
print(df.shape, df_ml.shape)
```

```python
print(df_ml.head())
```

```python
# делаем выборки
features = df_ml.drop(['num_orders'], axis=1) # выделение признаков
target = df_ml['num_orders']

features_train, features_test, target_train, target_test = train_test_split(
    features, target, shuffle=False, test_size=0.1, random_state=12345)
```

### Обучение моделей

#### Линейная регрессия

```python
%%time

# без масштабирования данных
# для кросс-валидации временных рядом используют tscv
tscv = TimeSeriesSplit(n_splits=9)
regressor = LinearRegression()
print('# Train LR for root_mean_squared_error #')
print()
cv_rmse_lr = (cross_val_score(regressor, 
                             features_train, 
                             target_train, 
                             cv=tscv, 
                             scoring='neg_mean_squared_error').mean() * -1) ** 0.5
print('Mean RMSE from CV of LinearRegression =', cv_rmse_lr)
```

#### DecisionTree

```python
%%time

regressor = DecisionTreeRegressor()
tscv = TimeSeriesSplit(n_splits=9)
# зададим перебор значений однострочными циклами
max_depth_list = [x for x in range(2, 15)]
min_samples_list = [x for x in range(1, 4)]
hyperparams = [{'criterion':['mse'], 
                'max_depth':max_depth_list,
                'min_samples_leaf':min_samples_list,
                'random_state':[12345]}]

print('# Tuning hyper-parameters DTR for root_mean_squared_error #')
print()
clf = GridSearchCV(regressor, hyperparams, scoring='neg_mean_squared_error', cv=tscv)
clf.fit(features_train, target_train)
print("Best parameters set found on development set:")
print()
print(clf.best_params_)
print()
print("Grid scores on development set:")
print()

# выводим результаты перебора
means_serial = clf.cv_results_['mean_test_score']
stds_serial = clf.cv_results_['std_test_score']
for mean, std, params in zip(means_serial, stds_serial, clf.cv_results_['params']):
    print("%0.2f for %r"% ((mean*-1)** 0.5, params))
print()

cv_rmse_dtr = (max(means_serial)*-1) ** 0.5
print('Mean RMSE from CV of DecissionTree regression =', cv_rmse_dtr)
print()
```

#### RandomForest

```python
%%time

regressor = RandomForestRegressor()
tscv = TimeSeriesSplit(n_splits=9)
# зададим перебор значений однострочными циклами
est_values_list = [x for x in range(50, 100, 10)]
max_depth_list = [x for x in range(2, 10)]
min_samples_list = [x for x in range(1, 4)]
hyperparams = [{'criterion':['mse'],
                'n_estimators':est_values_list,
                'max_depth':max_depth_list,
                'min_samples_leaf':min_samples_list,
                'random_state':[12345]}]

print('# Tuning hyper-parameters RF for root_mean_squared_error #')
print()
clf = GridSearchCV(regressor, hyperparams, scoring='neg_mean_squared_error', cv=tscv)
clf.fit(features_train, target_train)
print("Best parameters set found on development set:")
print()
print(clf.best_params_)
print()

cv_rmse_rf = (max(clf.cv_results_['mean_test_score'])*-1) ** 0.5
print('Mean RMSE from CV of RandomForest regression =', cv_rmse_rf)
print()
```

#### CatBoost

```python
%%time

regressor = CatBoostRegressor() 
tscv = TimeSeriesSplit(n_splits=9)
hyperparams = [{'learning_rate':[0.1, 0.2, 0.3],
                'iterations':range(100, 600, 100),
                'random_state':[12345],
                'verbose':[False]}]

print('# Tuning hyper-parameters CB for root_mean_squared_error')
print()
clf = GridSearchCV(regressor, hyperparams, scoring='neg_mean_squared_error', cv=tscv)
clf.fit(features_train, target_train)
print("Best parameters set found on development set:")
print()
print(clf.best_params_)
print()

cv_rmse_cb = (max(clf.cv_results_['mean_test_score'])*-1) ** 0.5
print('Mean RMSE from CV of CatBoost regression =', cv_rmse_cb)
print()
```

#### LGBM

```python
%%time

regressor = LGBMRegressor() 
tscv = TimeSeriesSplit(n_splits=9)
hyperparams = [{'learning_rate':[x/10 for x in range(1, 6)],
                'random_state':[12345]}]

print('# Tuning hyper-parameters LGBM for root_mean_squared_error')
print()
clf = GridSearchCV(regressor, hyperparams, scoring='neg_mean_squared_error', cv=tscv)
clf.fit(features_train, target_train)
print("Best parameters set found on development set:")
print()
print(clf.best_params_)
print()

cv_rmse_lgbm = (max(clf.cv_results_['mean_test_score'])*-1) ** 0.5
print('Mean RMSE from CV of LGBM regression =', cv_rmse_lgbm)
print()
```

#### Вывод  
Создали дополнительные признаки.  
Провели обучение 5 моделей c подбором и фиксацией наилучших гиперпараметров. Не забыли применять кросс-валидацию и учесть, что данные представляют собой временной ряд (`TimeSeriesSplit`).  
Наилучший результат продемонстрировала `CatBoost Regression` с RMSE = 23.834.

```python
%%time

model = CatBoostRegressor(learning_rate=0.1,
                          random_state=12345,
                          verbose=False) 
model.fit(features_train, target_train)
prediction = model.predict(features_test)
pred_rmse_lgbm = mean_squared_error(target_test, prediction, squared=False)
print("LGBM RMSE on the test samples =", pred_rmse_lgbm)

plt.figure(figsize=(15, 8))
plt.plot(target_test.index, 
         target_test, 
         'b-', 
         target_test.index, 
         prediction,
         'r-')
plt.show()
```
